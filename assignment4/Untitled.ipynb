```python
def distance(test_row, x, m):
    dist = 0
    for i in range(x.shape[1]):
        dist = dist + np.square(x[m][i]-test_row[i])
    dist = np.sqrt(dist)
    return dist
    
```

```python
print(distance(testset[3], x, 5))
```

```python
l = np.empty([x.shape[0],2])
for i in range(x.shape[0]):
    l[i][0] = distance(testset[0], x, i)
    l[i][1] = i
    print(distance(testset[0], x, i))
print(l)    
print(len(l))
print(l[0][0])
print('----------------------------------------------')
print(l[0:5,:])
print(l[0:5, :].shape)
```

```python
def give_list(test_row, x):
    l = np.empty([x.shape[0],2])
    for i in range(x.shape[0]):
        l[i][0] = distance(test_row, x, i)
        l[i][1] = i
    l = l[l[:,0].argsort()] 
    return l[0:5, :]
```

```python
def give_majority(l, y):
    m = l.shape[0]
    n_zero = 0
    n_one = 0
    for i in range(m):
        if y[int(l[i][1])]==0:
            n_zero = n_zero + 1
        else:
            n_one = n_one + 1
    if n_one > n_zero:
        return 1
    else:
        return 0
```

```python
p = give_list(testset[25], x)
print(p)
print(give_majority(p, y))
```

```python
def check_testset(testset, x, y):
    for i in range(testset.shape[0]):
        l = give_list(testset[i], x)
        print(give_majority(l, y))
        
```

```python
check_testset(testset, x, y)
```
