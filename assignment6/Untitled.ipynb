```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sig_der(x):
    return(sigmoid(x)*(1-sigmoid(x)))
```

```python
def give_sum(w, x):
    sums = np.zeros(x.shape[0],)
    for i in range(x.shape[0]):
        sums[i] = np.dot(w, x[i]) 
    return sums
```

```python
def give_class(sig):
    for i in range(sig.shape[0]):
        if sig[i] > 0.5:
            sig[i] = 1
        else:
            sig[i] = 0
    return sig        
```

```python
ex = give_sum(w, x_)
print(ex)
ex1 = sigmoid(ex)
print(ex1)
print(sigmoid(ex[0]))
print(y)
```

```python
#y_pred is the sigmoid as above
def give_error(y, y_pred):
    error = 0
    for i in range(y.shape[0]):
        error = error + (y[i] - y_pred[i])*(y[i] - y_pred[i])
    return 0.5*error
```

```python
print(give_error(ex1, y))
```

```python
def dE_dw(x_, y, y_pred, w):
    grad = np.zeros(w.shape[0],)
    for i in range(w.shape[0]):
        q = 0
        for j in range(x.shape[0]):
            q = q - (y[j] - y_pred[j])*y_pred[j]*(1 - y_pred[j])*x_[j][i]
        grad[i] = q
    return grad
```

```python
print(dE_dw(x_, y, ex1, w))
```

```python
print(w)
print(dE_dw(x_, y, ex1, w))
print( w - 0.1*dE_dw(x_, y, ex1, w))
```
